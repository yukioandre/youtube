```python
import pandas as pd
import numpy as np
```

```python
fifa19 = pd.read_csv("fifa19.csv")
fifa19
```

```python
# sintaxe: pd.pivot_table(dataset, index=, aggfunc=)
fifa19_filtered = fifa19[fifa19['Nationality'].isin(['Argentina', 'Brazil', 'France','Germany', 'Italy'])]
pd.pivot_table(fifa19_filtered, index='Nationality', aggfunc = np.mean)
```

```python
pd.pivot_table(fifa19_filtered, index='Nationality', aggfunc = np.median)
```

```python
pd.pivot_table(fifa19_filtered, index='Nationality', 
               values=['Age', 'Overall', 'Acceleration'],
               aggfunc = np.median)
```

```python
pivot_table = pd.pivot_table(fifa19_filtered, index=['Nationality','Club'], 
               values=['Age', 'Overall', 'Acceleration'],
               aggfunc = np.median)

pivot_table.to_csv("pivot_table.csv", sep=';', decimal=',')
```

```python
pd.pivot_table(fifa19_filtered, index=['Nationality','Club'], 
               values=['Age', 'Overall', 'Acceleration'],
               columns='Position',
               aggfunc = np.median)

```

```python
pd.pivot_table(fifa19_filtered, index=['Nationality'], 
               values=['Age', 'Overall', 'Acceleration'],
               columns='Club',
               aggfunc = np.median)
```
