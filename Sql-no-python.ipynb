```python
import pandas as pd
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
import pandasql as ps
```

```python
query = """select * from iris"""
```

```python
ps.sqldf(query, locals())
```

```python
query = """select Species, avg(SepalLengthCm) as avg_SepalLengthCm
from iris
group by Species
order by avg_SepalLengthCm desc """

ps.sqldf(query, locals())
```

```python
query = """select * from iris
where upper(Species) like '%SETOSA%'
"""

setosa = ps.sqldf(query, locals())
setosa.head()
```
